LINEAR REGRESSION

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
#sample data
X=np.array([1,2,3,4,5,6]).reshape(-1,1)#Featurre
y=np.array([2,4,6,8,10,12])#Target
```

```python
#Creating and fitting the model
model=LinearRegression()
model.fit(X,y)
```

```python
#make predictions
y_pred=model.predict(X)
```

```python
#Print model details
print('intercept: ', model.intercept_)
print('coeffiecient: ', model.coef_)
```

```python
#plot data and regression line
plt.scatter(X,y, color='blue', label='actual data')
plt.plot(X, y_pred, color='red', label='regression line')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.show()
```

POLYNOMIAL REGRESSION

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import Pipeline
```

```python
#creating polynomial regression pipeline
degree=3# using x^2 and x^3
poly_model=Pipeline([('poly_features', PolynomialFeatures(degree)),('lin_reg',LinearRegression())])
```

```python
#Fit the polynomial model
poly_model.fit(X,y)
```

```python
#performing predictions
X_range=np.linspace(1,5,100).reshape(-1,1)
y_poly_pred=poly_model.predict(X_range)
```

```python
#plotting the outputs
plt.scatter(X,y, color='blue', label='Actual Data')
plt.plot(X_range,y_poly_pred, color='purple', label='Polynomial Regression')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.show()
```

EVALUATION METRICES

```python
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np

mse = mean_squared_error(y, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y, y_pred)

print("MSE:", mse)
print("RMSE:", rmse)
print("R-squared:", r2)
```
